### Bond between CIM DOPO and QUBO

----
Каждый импульс DOPO в когерентной машине Изинга находится в состоянии сжатого вакуума, когда скорость накачки значительно ниже порога колебаний и в (сжатом) когерентном состоянии либо с 0-фазой, либо с π-фазой. В критической точке порога ДОФО существует линейная суперпозиция состояний 0-фазы и π-фазы
Синфазная и квадратурная амплитуды одиночного изолированного импульса DOPO подчиняются следующим c-числовым стохастическим дифференциальным уравнениям (CSDE):
$$\frac{{{\mathrm{d}}c_n}}{{{\mathrm{d}}t}} = \left( { - 1 + p - c_n^2 - s_n^2} \right)c_n + \frac{1}{{A_s}}\sqrt {c_n^2 + s_n^2 + \frac{1}{2}} \frac{{{\mathrm{d}}\zeta _n}}{{{\mathrm{d}}t}},$$
$$\frac{{{\mathrm{d}}s_n}}{{{\mathrm{d}}t}} = \left( { - 1 - p - c_n^2 - s_n^2} \right)s_n + \frac{1}{{A_s}}\sqrt {c_n^2 + s_n^2 + \frac{1}{2}} \frac{{{\mathrm{d}}\zeta _{n\prime }}}{{{\mathrm{d}}t}}.$$
Уравнения движения описывают временную эволюцию синфазной составляющей $c_n$ и квадратурной составляющей $s_n$ оптического поля n-й DOPO для сети из N импульсов DOPO, циркулирующих в кольцевом резонаторе. $p$ - параметр накачки, где p  = 1 представляет порог бифуркации вил и, таким образом, аналогичен силе обратной связи $α$ в CIM на основе OEO. $A_s$ - амплитуда насыщения при нормированной скорости накачки. Поле накачки адиабатически исключается в этих уравнениях, если предположить, что скорость затухания фотонов накачки намного больше, чем скорость распада сигнального фотона. $\frac{{{\mathrm{d}}\zeta }}{{{\mathrm{d}}t}}$ - является гауссовским термином белого шума. Вакуумные флуктуации сигнального канала вносят вклад в член 1/2, а квантовый шум поля накачки вносит вклад в $с^2+s^2$. 
Модель Изинга реализуется путем когерентного связывания импульсов DOPO. В CIM на основе измерения с обратной связью это достигается путем измерения синфазной составляющей каждого импульса $\tilde c_n[k]$ DOPO во время прохождения k туда и обратно со сбалансированным гомодинным обнаружением, а затем с использованием этого сигнала для генерации электронного сигнала обратной связи , который инжектируется обратно в полость путем модуляции последовательности оптических импульсов.
Для описания спинхфазной и квадратурной составляющей вблизи порогового значения накачки данные уравнения не действуют и имеют место квнтовые эффекты.


Интерес к представлению модели Изинга в терминах QUBO возник из-за возможности отображения гамильтониана как: $$\mathcal{H}_{Ising}=\underbrace{{-A(t)}\left(\sum_{i} \hat{\sigma}_{x}^{(i)}\right)}_{\text {Initial Hamiltonian }}+\underbrace{{B(t)}\left(\sum_{i} h_{i} \hat{\sigma}_{z}^{(i)}+\sum_{i,j} J_{i, j} \hat{\sigma}_{z}^{(i)} \hat{\sigma}_{z}^{(j)}\right)}_{\text {Final Hamiltonian }},$$ соответственно состояния спинов $\left| 0 \right\rangle$ и $\left| 1 \right\rangle$, где $B(t)$,$A(t)$ - являются функциями от времени, 
$t \in [0, T],\hat{\sigma}_{x, z}^{(i)}$ – операторами Паули, действующими на кубиты, а $h_i, J_{i,j}$ – локальными полями и взаимодействиями кубитов соответственно. 

Физически, вектор $h$ соответствует внешнему магнитному полю, применимому к каждому кубиту, матрица $J_{i,j}$ отвечает за попарные  взаимодействия кубитов. Для квантовых фазовых переходов второго рода верно, что $\Delta_{\text{min}} \propto N^{-l}$, что приводит к $T \propto N^{2l}$,то есть полиномиальному времени алгоритма.
Выражение выше описано в случае представления annealing schedule, возможно представить и изменение амплитуды спинов в случае моделирования спинов в сети с использованием DOPO. Тогда
$H_{FH}$ сопоставляется с эффективной потерей скорости распада фотонов сети DOPO путем реализации взаимного соединения, пропорционального $J_{ij}$ между двумя DOPO и инъекцией поля dc пропорционально $h_i$.


Основное состояние Ising Hamiltonian соответствует режиму колебаний с минимальной потерей сети. При скорости накачки ниже порога колебаний каждый DOPO находится в сжатом вакуумном состоянии, что интерпретируется как линейное суперпозиция $σ_i=+1$ (положительная амплитуда) и $σ_i =−1$ (отрицательная амплитуда), так что вероятность нахождения амплитуды вращения в одном из этих состояний для всех $2^N$ конфигураций идентичны. Когда усиление сети DOPO увеличивается, чтобы достичь минимальной скорости потери, в основном состоянии происходит одномодовое колебание. 

Входные состояния фотонов описываются $\left| 0 \right\rangle _1\left| 0 \right\rangle _2, \cdots \left| 0 \right\rangle _N$. Приложенное пороговое значение усилиния переводит все связные состояния в положительную амплитуду |α⟩
или отрицательную |−α⟩.

Возможно представить функцию усиления нейронов в классических нейронных сетях ( CNN )
Предложено уравнение для описания состояния, в нейтральном состоянии выражается непрерывной переменной $a_i$, что подчиняется непрерывной эволюции времени, управляемой:
$$\frac{{\rm d}}{{{\rm d}t}}a_i = - a_i + f\left( {a_i} \right) - \frac{{\partial V}}{{\partial a_i}} + g_i,$$
где первый член выражает линейную потерю, а второй член описывает самообращение с нелинейной функцией усиления f, время нормализуется скоростью затухания возбуждения нейрона. Третий член представляет собой взаимную связь между нейронами и потенциалом взаимодействия V, что реализует заданную задачу, $g_i$ - выражает движущую силу шума, связанную с распадом и нелинейным усилением для возбуждения нейрона.

## Подход к решению MaxCut
Пусть задан граф на множестве вершин V, связанных множеством ребер E. Каждое ребро имеет две инцидентных вершины w,u. Также заадны веса для ребер $\in$ {-1,1}. В общем случае нахождение 
В общем случае данная задача представляет минимизацию функции весов 
$$C = \sum_{u,v,w \in E} w (\mathbf{1}(u \in V_1, v \in V_2) + \mathbf{1}(u \in V_2, v \in V_1))$$
Используя квантовые вычисления в реализации поиска основного состояния полного взвешеного графа занимает полиноминальное время и является NP-hard, поиск k-го разреза графа занимает $O(m^2n+2m^mn)$ время, где n-k -слабосвязные вершины, а m - ребер для n-ой вершины.
Однако время решения данной проблемы растет экспоненциально $O(N) \sim e^{N}$ и нахождение точного решения для данной проблемы возможно в случае малого числа вершин (пример 1). 

Ограничения в использовании для решения D-Wave Ocean SDK (DWave Leap): при использовании BQM and QOBO (based on QPU) - 109 клик и 5096 кубит. Для Qiskit IBM Lab (SamplingVQE) - ограничение в 64 кубита, каждый из которых представляет собой бра-кет нотацию элемента QOBO.

В случае с графом из 10 связанных между собой вершин возможно нахождение решения.
Для системы из большего числа вершин: предложенное решение не является достоверным.

```python
import random
import numpy as np
import networkx as nx
import dwave_networkx as dnx
from dwave.system.samplers import DWaveSampler
from dwave.system.composites import EmbeddingComposite
from matplotlib import pyplot as plt
from collections import defaultdict
n = 20  # number of nodes
G = nx.complete_graph(n)
h = defaultdict(int)
J = defaultdict(int)

random.seed(451)
for (u, v) in G.edges():
    J[(u, v)] = 2*np.random.randint(-1, 1)-1

numruns = 50
sampler = EmbeddingComposite(DWaveSampler())
response = sampler.sample_ising(h, J,
                                num_reads=numruns,
                                label='Maximum Cut Ising')
print('Set -1', 'Set 1', 'Energy', 'Cut Size', sep=' ')
for sample, E in response.data(fields=['sample', 'energy']):
    UP = [k for k, v in sample.items() if v == -1]
    D = [k for k, v in sample.items() if v == 1]
    print(str(UP), str(D), str(E), str(int((20-E)/2)), sep=' ')
lut = response.first.sample
# Interpret best result in terms of nodes and edges
U = [node for node in G.nodes if lut[node] == -1]
DN = [node for node in G.nodes if lut[node] == 1]
cut_edges = [(u, v) for u, v in G.edges if lut[u] != lut[v]]
uncut_edges = [(u, v) for u, v in G.edges if lut[u] == lut[v]]

# Display best result
pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos, nodelist=U, node_color='r')
nx.draw_networkx_nodes(G, pos, nodelist=DN, node_color='c')
nx.draw_networkx_edges(G, pos, edgelist=cut_edges, style='dashdot', alpha=0.5,
                       width=3)
nx.draw_networkx_edges(G, pos, edgelist=uncut_edges, style='solid', width=3)
plt.suptitle(f"Best cat:\
     with energy:\
    {round(max(response.data(fields=['energy']))[0],3)}")
nx.draw_networkx_labels(G, pos)


filename = "maxcut_20_node.png"
plt.savefig(filename, bbox_inches='tight', )
```
